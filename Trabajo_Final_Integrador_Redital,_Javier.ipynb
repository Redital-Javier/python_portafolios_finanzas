###  Parte 1.2: Descarga de datos histricos

Se descargan los precios ajustados de cierre para los 12 activos seleccionados desde **Yahoo Finance**, cubriendo el perodo comprendido entre **enero de 2022** y **junio de 2025**.

La fuente de datos ser el mdulo `yfinance`, el cual permite acceder directamente a datos de mercado desde Python.


```python
# Paso 1: Instalar y actualizar yfinance si es necesario
!pip install yfinance --upgrade --quiet

# Paso 2: Importar libreras necesarias
import yfinance as yf
import pandas as pd

# Paso 3: Definir los 12 tickers seleccionados
tickers = ['SPY', 'BRK-B', 'XLP', 'AAPL', 'GOOG', 'KO', 'MELI', 'NVDA', 'MA', 'YPF', 'BTC-USD', 'PFE']

# Paso 4: Descargar los precios ajustados diarios entre enero 2022 y junio 2025
data = yf.download(tickers, start='2022-01-01', end='2025-06-30')

# Paso 5: Seleccionar la columna de precios de cierre
data = data["Close"]

# Paso 6: Eliminar fechas con valores faltantes en algn activo
data = data.dropna()

# Paso 7: Mostrar las primeras filas del DataFrame resultante
data.head()
```

###  Parte 2.1: Evolucin de precios histricos

Se grafican los precios ajustados de los activos para observar su evolucin entre enero de 2022 y junio de 2025.

La visualizacin se realiza en dos etapas complementarias:

1. **Grficos individuales:** se representa cada activo por separado para identificar su comportamiento propio (tendencias, rupturas, niveles de volatilidad, etc.).

2. **Grfico comparado (base 100):** se normalizan todos los precios a una misma base inicial (100) y se grafican juntos en un solo panel. Esto permite comparar el rendimiento relativo de cada instrumento a lo largo del tiempo, sin importar el valor nominal de sus precios.

Estos grficos permiten tener una visin clara del desempeo individual y colectivo de los activos seleccionados.


```python
# Paso 1: Grficos individuales
# Se genera un grfico por cada activo para observar su comportamiento de forma separada.

import matplotlib.pyplot as plt

# Configura el tamao general y el diseo de subplots
n = len(data.columns)
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(18, 12), sharex=True)
axes = axes.flatten()

# Genera un grfico por cada activo
for i, ticker in enumerate(data.columns):
    axes[i].plot(data.index, data[ticker], lw=1.5)
    axes[i].set_title(f"{ticker}", fontsize=10)
    axes[i].grid(True)

# Ajusta el layout para que no se superpongan elementos
plt.suptitle("Evolucin individual de precios ajustados (20222025)", fontsize=14)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```

```python
# Paso 2: Grfico comparado (base 100)
# Se normalizan los precios para que todos comiencen desde 100 y se pueda comparar su evolucin relativa.

normalized_data = data / data.iloc[0] * 100

# Crear el grfico
plt.figure(figsize=(14, 7))
for ticker in normalized_data.columns:
    plt.plot(normalized_data.index, normalized_data[ticker], label=ticker, lw=1.5)

plt.title("Evolucin comparada de precios ajustados (20222025)\nTodos comienzan desde 100")
plt.xlabel("Fecha")
plt.ylabel("ndice de precio (base 100)")
plt.legend(loc="upper left", ncol=3, fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Parte 3.1: CAPM y Betas

- iguales a 1: se mueven en lnea con el mercado  
- menores a 1: son menos voltiles que el mercado

- `rf` = tasa libre de riesgo  
- `` = coeficiente estimado  mide la sensibilidad al mercado


```python
import statsmodels.api as sm

# Tasa libre de riesgo diaria (usamos la misma que antes)
rf_daily = 0.04 / 252

# Calculamos retornos excedentes (activo - rf)
excess_returns = data_returns.sub(rf_daily)

# Retornos del mercado (SPY)
market_excess = data_returns["SPY"] - rf_daily

# Estimamos beta para cada activo (excepto SPY)
betas = {}

for ticker in data_returns.columns:
    if ticker != "SPY":
        y = excess_returns[ticker]
        x = market_excess
        x = sm.add_constant(x)
        modelo = sm.OLS(y, x).fit()
        betas[ticker] = modelo.params[1]  # Coeficiente beta (pendiente)

# Convertimos a DataFrame ordenado
df_betas = pd.Series(betas).sort_values().to_frame(name="Beta").round(4)
df_betas
```

###  Parte 3.2: Clculo de ratios de performance: Sharpe, Sortino y Treynor

Para evaluar el desempeo ajustado por riesgo de cada activo, se calculan tres ratios financieros fundamentales:

- **Ratio de Sharpe**: mide el exceso de retorno sobre la tasa libre de riesgo por unidad de volatilidad total. Es til para comparar activos con distinto nivel de riesgo.

- **Ratio de Sortino**: similar al de Sharpe, pero solo penaliza la volatilidad negativa (cadas), lo cual lo vuelve ms preciso para evaluar inversiones de perfil moderado.

- **Ratio de Treynor**: relaciona el exceso de retorno con el riesgo sistemtico (beta). Es adecuado para carteras bien diversificadas, donde solo importa el riesgo de mercado.

Los valores se anualizan y permiten identificar cules activos ofrecen mejor retorno ajustado por el tipo de riesgo que representan.


```python
import numpy as np

# Diccionario para guardar los ratios
ratios = {}

# Calcular ratios para todos los activos excepto SPY
for ticker in data_returns.columns:
    if ticker != "SPY":
        r = data_returns[ticker]
        ex_r = r - rf_daily  # Exceso de retorno diario

        # Sharpe Ratio
        sharpe = (ex_r.mean() / ex_r.std()) * np.sqrt(252)

        # Sortino Ratio
        downside_std = r[r < rf_daily].std()
        sortino = (ex_r.mean() / downside_std) * np.sqrt(252)

        # Treynor Ratio
        treynor = (ex_r.mean() * 252) / betas[ticker]

        # Guardar resultados
        ratios[ticker] = {
            "Sharpe": sharpe,
            "Sortino": sortino,
            "Treynor": treynor
        }

# Crear tabla ordenada y redondeada
ratios_df = pd.DataFrame(ratios).T.round(2)
ratios_df.sort_values(by="Sharpe", ascending=False)
```

###  Parte 3.3: Regresin con el Modelo Fama-French de 3 Factores

Para mejorar la explicacin del retorno de los activos, se aplica el **modelo Fama-French de tres factores**, una extensin del CAPM que considera no solo el riesgo de mercado, sino tambin el efecto del tamao de la empresa y su valoracin.

**Frmula:**  
**R - Rf =  +   (Rm - Rf) +   SMB +   HML + error**

 Qu significan los factores?

- **Rm - Rf**: Exceso de retorno del mercado (igual que en el CAPM).
- **SMB (Small Minus Big)**: mide la diferencia de rendimiento entre acciones de empresas pequeas y grandes. Captura el efecto de tamao.
- **HML (High Minus Low)**: mide la diferencia entre acciones con alto valor contable/mercado y aquellas con bajo. Captura el efecto de "value vs. growth".
- ****: el rendimiento adicional que no puede explicarse por los tres factores (alpha).
- **, , **: sensibilidad de la accin a cada factor.

Este modelo permite evaluar si un activo tiene rendimientos anormales (  0) o si sus retornos pueden explicarse completamente por estos factores sistemticos.


```python
# Descarga y preparacin de los factores (diarios) desde la Kenneth French Data Library
url = "https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_Factors_daily_CSV.zip"
factors = pd.read_csv(url, skiprows=3)
factors = factors.rename(columns={"Unnamed: 0": "Date"})
factors = factors[factors["Date"].str.isnumeric()]
factors["Date"] = pd.to_datetime(factors["Date"], format="%Y%m%d")
factors = factors.set_index("Date").astype(float) / 100  # de % a decimal
factors = factors[["Mkt-RF", "SMB", "HML", "RF"]]

# Visualizamos las primeras filas del DataFrame ya limpio
factors
```

```python
# Unin del DataFrame de factores con los retornos diarios

# Se combinan los factores Fama-French con los retornos diarios de los activos
# Solo se conservan las fechas en comn (interseccin de ambos ndices)
data_ff = data_returns.join(factors, how="inner")

import statsmodels.api as sm

# Diccionario vaco para guardar los resultados de cada activo
ff_results = {}

# Se calcula el modelo Fama-French para cada activo del portafolio
for ticker in data_returns.columns:
    # Exceso de retorno del activo: Ri - Rf
    Ri = data_ff[ticker] - data_ff["RF"]

    # Variables independientes: factores del modelo
    X = data_ff[["Mkt-RF", "SMB", "HML"]]
    X = sm.add_constant(X)  # agrega constante para estimar alpha

    # Regresin lineal OLS
    modelo = sm.OLS(Ri, X).fit()

    # Se guardan los coeficientes y el R ajustado
    ff_results[ticker] = {
        "Alpha": round(modelo.params["const"], 4),
        "Beta_Mkt": round(modelo.params["Mkt-RF"], 4),
        "Beta_SMB": round(modelo.params["SMB"], 4),
        "Beta_HML": round(modelo.params["HML"], 4),
        "R2_ajustado": round(modelo.rsquared_adj, 4)
    }

# Mostrar la tabla resumen con los resultados
pd.DataFrame(ff_results).T
```

##  Parte 4.1: Identificacin del Portafolio de Mxima Sharpe

En esta seccin se analizar una estrategia que prioriza la **eficiencia en el uso del riesgo**, buscando maximizar el retorno esperado por cada unidad de volatilidad asumida.

El objetivo es identificar el portafolio que presenta el **mayor ratio de Sharpe** entre las 10.000 combinaciones simuladas. Esta mtrica es ampliamente utilizada en finanzas para evaluar la calidad de una estrategia de inversin, ya que ajusta el retorno en funcin del riesgo total.

Una vez identificado, se analizar su rendimiento, composicin y posicin dentro de la frontera eficiente, permitiendo evaluar su conveniencia para un inversor con tolerancia media al riesgo y objetivos de crecimiento sostenido.

Este enfoque representa una estrategia balanceada pero orientada al rendimiento, ideal para perfiles moderados que buscan **optimizar su rentabilidad sin desprotegerse del riesgo**.


```python
# Fijar semilla para reproducibilidad
np.random.seed(42)

# Establecemos nmero de simulaciones
num_portafolios = 10000

# Calculamos medias y matriz de covarianza
media_retornos = data_returns.mean() * 252        # retorno anualizado
matriz_cov = data_returns.cov() * 252             # covarianza anualizada
tickers = data_returns.columns

# Inicializamos listas para guardar resultados
resultados = {
    "Retorno": [],
    "Volatilidad": [],
    "Sharpe": [],
    "Pesos": []
}

# Simulacin de portafolios aleatorios
for _ in range(num_portafolios):
    pesos = np.random.random(len(tickers))
    pesos /= np.sum(pesos)  # normalizamos

    retorno = np.dot(pesos, media_retornos)
    volatilidad = np.sqrt(np.dot(pesos.T, np.dot(matriz_cov, pesos)))
    sharpe_ratio = retorno / volatilidad

    resultados["Retorno"].append(retorno)
    resultados["Volatilidad"].append(volatilidad)
    resultados["Sharpe"].append(sharpe_ratio)
    resultados["Pesos"].append(pesos)

# Convertimos resultados a DataFrame
portafolios = pd.DataFrame(resultados)
portafolios["Pesos"] = portafolios["Pesos"].apply(lambda x: np.round(x, 4))

# Mostramos primeras filas
portafolios.head()
```

##  Parte 4.2: Identificacin del Portafolio de Mnima Volatilidad

En esta seccin se analizar una estrategia alternativa al enfoque de mxima eficiencia (Sharpe), priorizando la **estabilidad** del portafolio por sobre la rentabilidad.

El objetivo es identificar el portafolio que presenta la **menor volatilidad anualizada** entre las 10.000 combinaciones simuladas. Esta estrategia puede ser especialmente relevante para inversores con una alta aversin al riesgo o con un horizonte de inversin conservador.

Una vez identificado, se analizar su rendimiento esperado, su composicin y se comparar visualmente con el resto de las combinaciones simuladas.

Este enfoque permite evaluar una estrategia basada en la **minimizacin del riesgo absoluto**, lo cual tambin puede ser compatible con un perfil de inversor moderado en ciertos contextos de mercado.


```python
# Fijar semilla para reproducibilidad
np.random.seed(42)

# Establecer nmero de simulaciones
num_portafolios = 10000

# Calcular medias y matriz de covarianza
media_retornos = data_returns.mean() * 252         # retorno anualizado
matriz_cov = data_returns.cov() * 252              # covarianza anualizada
tickers = data_returns.columns                     # nombres de los activos

# Inicializar listas para guardar resultados
resultados = {
    "Retorno": [],
    "Volatilidad": [],
    "Sharpe": [],
    "Pesos": []
}

# Simulacin de portafolios aleatorios
for _ in range(num_portafolios):
    pesos = np.random.random(len(tickers))
    pesos /= np.sum(pesos)  # normalizar los pesos para que sumen 1

    retorno = np.dot(pesos, media_retornos)
    volatilidad = np.sqrt(np.dot(pesos.T, np.dot(matriz_cov, pesos)))
    sharpe_ratio = retorno / volatilidad

    resultados["Retorno"].append(retorno)
    resultados["Volatilidad"].append(volatilidad)
    resultados["Sharpe"].append(sharpe_ratio)
    resultados["Pesos"].append(pesos)

# Convertir a DataFrame y redondear los pesos
portafolios = pd.DataFrame(resultados)
portafolios["Pesos"] = portafolios["Pesos"].apply(lambda x: np.round(x, 4))

# Mostrar las primeras filas
portafolios.head()
```

```python
# Identificamos el portafolio con menor volatilidad
idx_min_vol = portafolios["Volatilidad"].idxmin()
min_vol_portafolio = portafolios.loc[idx_min_vol]

# Mostramos sus mtricas principales
min_vol_portafolio[["Retorno", "Volatilidad", "Sharpe"]]
```

```python
# Grfico de portafolios simulados
plt.figure(figsize=(10, 6))
scatter = plt.scatter(portafolios["Volatilidad"], portafolios["Retorno"],
                      c=portafolios["Sharpe"], cmap="viridis", alpha=0.6)
plt.colorbar(scatter, label="Ratio de Sharpe")

# Marcamos el portafolio de mnima volatilidad
plt.scatter(min_vol_portafolio["Volatilidad"], min_vol_portafolio["Retorno"],
            color="blue", marker="*", s=200, label="Mnima Volatilidad")

# Esttica
plt.title("Frontera eficiente  Portafolio de mnima volatilidad", fontsize=14)
plt.xlabel("Volatilidad anual")
plt.ylabel("Retorno anual esperado")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

##  Parte 4.3: Comparacin de la Evolucin del Rendimiento Acumulado

A continuacin se presenta un grfico comparativo entre los portafolios de **Mxima Sharpe** y **Mnima Volatilidad**, que permite observar visualmente cmo habra evolucionado una inversin en cada estrategia a lo largo del tiempo (20222025).

El anlisis se basa en la simulacin de rendimiento diario acumulado, partiendo de una inversin inicial equivalente a \$1.



```python
# Convertir pesos definitivos a arrays de numpy
pesos_sharpe_array = np.array([
    0.2553,  # YPF
    0.1671,  # NVDA
    0.1769,  # KO
    0.1032,  # BTC-USD
    0.0234,  # XLP
    0.0165,  # SPY
    0.0226,  # PFE
    0.0014,  # MELI
    0.0445,  # AAPL
    0.1183,  # BRK-B
    0.0321,  # MA
    0.0388   # GOOG
])

pesos_vol_array = np.array([
    0.0008,  # YPF
    0.0165,  # NVDA
    0.0745,  # KO
    0.0802,  # BTC-USD
    0.2407,  # XLP
    0.1344,  # SPY
    0.2202,  # PFE
    0.0054,  # MELI
    0.0018,  # AAPL
    0.0365,  # BRK-B
    0.14,    # MA
    0.0489   # GOOG
])

# Reordenar columnas de retornos para que coincidan con el orden de los pesos
orden_tickers = ["YPF", "NVDA", "KO", "BTC-USD", "XLP", "SPY",
                 "PFE", "MELI", "AAPL", "BRK-B", "MA", "GOOG"]
data_returns_ordenado = data_returns[orden_tickers]

# Retornos diarios de cada portafolio
retorno_diario_sharpe = data_returns_ordenado.dot(pesos_sharpe_array)
retorno_diario_vol = data_returns_ordenado.dot(pesos_vol_array)

# Retorno acumulado (multiplicativo)
retorno_acum_sharpe = (1 + retorno_diario_sharpe).cumprod()
retorno_acum_vol = (1 + retorno_diario_vol).cumprod()

# Grfico final
plt.figure(figsize=(10, 6))
plt.plot(retorno_acum_sharpe, label="Mxima Sharpe", linewidth=2)
plt.plot(retorno_acum_vol, label="Mnima Volatilidad", linewidth=2)
plt.title("Evolucin del Rendimiento Acumulado de los Portafolios", fontsize=14)
plt.xlabel("Fecha")
plt.ylabel("Crecimiento del capital (Base = 1)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Mostrar los valores finales del rendimiento acumulado para cada portafolio
valor_final_sharpe = retorno_acum_sharpe.iloc[-1]
valor_final_vol = retorno_acum_vol.iloc[-1]

# Calcular el rendimiento acumulado total en %
retorno_acumulado_sharpe = (valor_final_sharpe - 1) * 100
retorno_acumulado_vol = (valor_final_vol - 1) * 100

# Mostrar resultados
print(f"Retorno acumulado total - Mxima Sharpe: {retorno_acumulado_sharpe:.2f}%  Capital final: ${valor_final_sharpe:.2f}")
print(f"Retorno acumulado total - Mnima Volatilidad: {retorno_acumulado_vol:.2f}%  Capital final: ${valor_final_vol:.2f}")
```



El grfico anterior permite visualizar de forma clara cmo evolucionaron en el tiempo los dos portafolios ptimos construidos mediante simulacin:

- El portafolio de **Mxima Sharpe** presenta una trayectoria ms voltil pero con un rendimiento acumulado significativamente superior. Esto refleja una estrategia orientada a **maximizar el retorno ajustado por riesgo**, an a costa de asumir mayor variabilidad.

- El portafolio de **Mnima Volatilidad**, en cambio, sigue una curva ms suave y estable, con un crecimiento moderado y consistente en el tiempo. Esta estrategia prioriza la **preservacin de capital y estabilidad**, lo cual puede ser ms adecuado para perfiles **moderados con baja tolerancia al riesgo**.

Ambas estrategias resultan viables para un inversor de perfil moderado, dependiendo de sus preferencias especficas de riesgo y objetivos de inversin. Este anlisis comparativo permite ilustrar con claridad las diferencias prcticas entre ambos enfoques.
